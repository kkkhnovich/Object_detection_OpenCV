```python
import cv2
import numpy as np
import timeit
```

```python
src = cv2.imread("cars.png")
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
car_classifier = cv2.CascadeClassifier('Resources/haarcascade_car.xml')
cars = car_classifier.detectMultiScale(gray, 1.1, 1)
```

```python
for (x,y,w,h) in cars:
        cv2.rectangle(src, (x, y), (x+w, y+h), (0, 255, 255), 2)
# Displaying the processed image on the screen
cv2.imshow('Cars', src)
if cv2.waitKey(0):
    cv2.destroyAllWindows()
```
